```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
movies = pd.read_csv('data/IMDb movies.csv', low_memory = False)
```

```python
# Distribution of the 10 Title Types (bar chart)
cmap = plt.get_cmap('tab10')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]

movie_genres = movies['genre'].str.split(',', expand=True)

ax = movie_genres.value_counts().head(15).plot(kind='barh', figsize=(20,10), color=colors)
# color=(0.2, 0.4, 0.6, 0.6),
plt.title('Bar Chart - Number of Titles by IMDb genres')
```
